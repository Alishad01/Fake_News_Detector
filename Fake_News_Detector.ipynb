```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

def generate_headline(prompt="Breaking:"):
    result = generator(prompt, max_new_tokens=30, do_sample=True, top_k=50, temperature=0.9)
    return result[0]['generated_text']
```

```python
print("🔹 Generated Headline:")
print(generate_headline("Breaking:"))
```

```python
from transformers import pipeline

classifier = pipeline("text-classification", model="nlptown/bert-base-multilingual-uncased-sentiment")

def detect_fake_news(text):
    result = classifier(text)[0]
    label = result["label"]
    score = result["score"]
    return f"Prediction: {label} with confidence {score:.2f}"
```

```python
print("\n🔹 Fake News Prediction:")
print(detect_fake_news("Aliens just landed in New York and opened a food truck."))
```
